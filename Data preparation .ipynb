```python
# convert alpha-2 code to alpha-3 code
hotel['country'].replace('CN', 'CHN', inplace=True)

# add country East timor (TMP) to continent dataset
TMP = {'name':'East Timor', 'alpha-3':'TMP', 'sub-region':'South-eastern Asia' }
continent = continent.append(TMP, ignore_index=True)
# add sub-region ATA to continent dataset
row = int(np.where(continent['alpha-3']=='ATA')[0])
continent.iloc[row,6] = 'Antarctica' 
```

```python
# add regions to hotel dataset
regions = []
n = 0
for code in hotel['country']:
    # if code == code is False, it is a NaN value
    if (code == code):
        #print(n)
        index = int(np.where(continent['alpha-3'] == code)[0])
        regions.append(continent['sub-region'][index])
    else:
        regions.append(np.nan)
    n += 1
    
hotel['sub_regions'] = regions
hotel.head()
```

```python
# split data
X, y = hotel3.iloc[:,1:], hotel3.iloc[:,0]
total_data = len(X.index)
split = round(total_data*0.8)
X_train = X.iloc[:split, :]
X_test = X.iloc[split:, :]
y_train = y.iloc[:split]
y_test = y.iloc[split:]
```

```python
# Data Normalization
# https://machinelearningmastery.com/data-preparation-without-data-leakage/
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
#X_test = scaler.transform(X_test_train)
```

```python
# Imputation (feature = agent)
#https://machinelearningmastery.com/knn-imputation-for-missing-values-in-machine-learning/
imputer = KNNImputer()
imputer.fit(X_train)
X_train = imputer.transform(X_train)
#X_test_trans = imputer.transform(X_test)
```

```python
print('missing: %d' % sum(np.isnan(X_train).flatten()))
```

### Below cell still need to be completed

```python
dtrain = xgb.DMatrix(data=X_train, label=y_train, feature_names=X.columns.tolist())
```

### Training 

```python
# initial model with cross_validation
# code found here: https://towardsdatascience.com/cross-validation-and-hyperparameter-tuning-how-to-optimise-your-machine-learning-model-13f005af9d7d
clf = xgb.XGBClassifier()
xgb_param = clf.get_xgb_params()
cvresult = xgb.cv(
                xgb_param, 
                dtrain, 
                num_boost_round=clf.get_params()['n_estimators'], 
                nfold=5,
                metrics=['auc', 'error'], 
                early_stopping_rounds=50,
                verbose_eval=10)
```

```python
train_error = cvresult['train-error-mean']
test_error = cvresult['test-error-mean']
epoch = len(train_error)
plt.plot(range(0,epoch), train_error, '-.b', label='train error')
plt.plot(range(0,epoch), test_error, '-r', label='test error')
plt.legend()
```

```python
#Fit the algorithm on the data
clf.fit(X_train, y_train, eval_metric='auc')

#Predict training set:
y_train_preds = clf.predict(X_train)
y_train_predprob = clf.predict_proba(X_train)[:,1]
```

```python
#Print model report:
print("\nModel Report")
print("Accuracy : %.4g" % accuracy_score(y_train, y_train_preds))
print("AUC Score (Train): %f" % roc_auc_score(y_train, y_train_preds))
print()
print('Classification Report (Train)')
print(classification_report(y_train, y_train_preds, target_names=['not_canceled', 'canceled']))
```
